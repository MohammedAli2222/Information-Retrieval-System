```python
# --- الخلية 1: اختبار خدمة معالجة وتحسين الاستعلامات ---
from services.query_refinement_service import QueryRefinementService

print("---  اختبار خدمة تحسين الاستعلام (Query Refinement) ---")
refiner = QueryRefinementService()


bad_query = "lung inflamation therpy"

print(f"الاستعلام الأصلي (قبل المعالجة): '{bad_query}'")


refined_query = refiner.refine_query(bad_query, use_spell_check=True, use_synonyms=True)

print(f"الاستعلام المحسن (بعد المعالجة): '{refined_query}'")
print(" النتيجة: تم تصحيح الأخطاء وإضافة المرادفات بنجاح!")
```

```python
# --- الخلية 2: اختبار خدمة تنظيف النصوص ---
from services.preprocessing_service import preprocess

print("\n---  اختبار خدمة معالجة النصوص (Preprocessing) ---")

raw_medical_text = "The patients are exhibiting severe respiratory symptoms and a high fever."
print(f"النص الأصلي:\n{raw_medical_text}\n")

extracted_tokens = preprocess(raw_medical_text)

print(f"الكلمات المستخرجة (Tokens) بعد التنظيف والتجذير:\n{extracted_tokens}")
print(" النتيجة: تم إزالة الكلمات الموقوفة (The, are, and) وتم إرجاع الكلمات لجذورها بنجاح!")
```

```python
# --- الخلية 3: اختبار خدمة التخزين السريع ---
import os
from services.cache_service import save_cache, load_cache

print("\n---  اختبار خدمة التخزين المؤقت (Cache Service) ---")

mock_data = {"doc_1": "Medical Abstract A", "doc_2": "Medical Abstract B"}
test_cache_filename = "test_memory.pkl"

print("1. جاري حفظ البيانات في الذاكرة المؤقتة...")
save_cache(mock_data, test_cache_filename)


print("2. جاري استرجاع البيانات من الذاكرة المؤقتة...")
loaded_data = load_cache(test_cache_filename)

if loaded_data == mock_data:
    print(f"البيانات المسترجعة: {loaded_data}")
    print(" النتيجة: عملية الحفظ والاسترجاع (I/O) تعمل بكفاءة تامة!")
else:
    print(" النتيجة: حدث خطأ في الاسترجاع.")
    

if os.path.exists(f"cache/{test_cache_filename}"):
    os.remove(f"cache/{test_cache_filename}")
```
